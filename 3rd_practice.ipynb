Импортируем библиотеки

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
import cv2
```

Инициализируем модель обучения VGG16

```python
model = keras.applications.VGG16(
    include_top=True,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classes=1000,
    classifier_activation="softmax",
)
```

Загружаем фотографию кота размером 224х224 пикселя

```python
cat_img=cv2.imread("shocked_cat.jpg")
plt.imshow(cat_img)
```

Приводим изображение к входному формату для модели

```python
cat_img = np.array(cat_img)
x = keras.applications.vgg16.preprocess_input(cat_img)
print(x.shape)
x = np.expand_dims(x, axis=0)
```

Проверяем нашу модель на тестовом изображении

```python
res = model.predict(x)
print(np.argmax(res))
```
