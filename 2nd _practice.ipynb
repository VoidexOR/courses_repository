Импортируем библиотеки и делаем первоначальные настройки

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

Создаём 2 массива, содержащие нужную для нас выборку

```python
c = np.array([-40, -10, 0, 8, 15, 22, 38]) # градусы Цельсия
f = np.array([-40, 14, 32, 46, 59, 72, 100]) # соотвествующие градусы Фаренгейта
```

Определяем модель нейронной сети и её структуру

```python
model = keras.Sequential()
model.add(Dense(units=1, input_shape=(1,), activation='linear'))
```

Компилируем её и запускаем процесс обучения

```python
model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(0.1))
history = model.fit(c, f, epochs=500, verbose=0)
print('Обучение прошло успешно!')
```

Просмотрим изменение количества потерь по графику обучения

```python
plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```

Смотрим на результат обучения нашей модели и образовавшийся весовой коэффициент

```python
print(model.predict([100]))
print(model.get_weights())
```
